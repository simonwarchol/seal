```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/Users/swarchol/Research/seal/data/updated_renamed_with_hdbscan.csv')

df['UMAP_Y'] = df['UMAP_Y'] * -1

```

```python
# Create a high resolution plot
plt.figure(figsize=(4, 4), dpi=300)  # Increased figure size and DPI

# Create the scatter plot
sns.scatterplot(
    x='UMAP_X', 
    y='UMAP_Y', 
    data=df,
    s=5,  # Adjust point size
    # alpha=0.6,  # Add some transparency
    rasterized=True,  # Rasterize the scatter points for better performance with high DPI,
    # make dots purple,
    color='purple'
)

# Remove axis values and ticks
plt.xticks([])
plt.yticks([])
# plt.axis('off')  # This removes the axes completely

# Add title with larger font
plt.title('2D Embedding Scatterplot')

# Adjust layout
plt.tight_layout()

# Save and display at high resolution
plt.savefig('embedding.png', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
