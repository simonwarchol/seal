```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_parquet('https://seal-vis.s3.us-east-1.amazonaws.com/WD-76845-097/df.parquet')

```

```python
# df.head()



df["UMAP_X"] = df["UMAP_X"]
df["UMAP_Y"] = df["UMAP_Y"].max() - df["UMAP_Y"]

# seaborn plot scatter UMAP_X_plot, UMAP_Y_plot
sns.scatterplot(x="UMAP_X", y="UMAP_Y", data=df, alpha=0.01)
plt.show()
```
